```sql
CREATE FUNCTION TRY_CONVERT_TO_UID (@uuid NVARCHAR(50))
RETURNS UNIQUEIDENTIFIER 
AS 
BEGIN
RETURN (
    SELECT  TRY_CAST(SUBSTRING(@uuid, 1, 8) + '-' + SUBSTRING(@uuid, 9, 4) + '-' + SUBSTRING(@uuid, 13, 4) + '-'+ SUBSTRING(@uuid, 17, 4) + '-' + SUBSTRING(@uuid, 21, 12) 
        AS UNIQUEIDENTIFIER)
)
END
```

```sql
WITH numberOfOrders (customer_unique_id, numberOfOrders)
AS (
SELECT customer_unique_id, COUNT(customer_id) AS numberOfOrders
FROM dbo.olist_customers_dataset
GROUP BY customer_unique_id
)
SELECT TOP 100 *
FROM dbo.olist_customers_dataset c
JOIN numberOfOrders nm ON  c.customer_unique_id = nm.customer_unique_id
WHERE numberOfOrders >= 2
ORDER BY c.customer_unique_id ;
```
